```python
import pickle
import pandas as pd
from pathlib import Path

from tensorflow.keras.preprocessing.text import Tokenizer

print("🚀 HelpDesk AI - Tokenizer Started")
```

```python
PROJECT_ROOT = Path.cwd().parent

PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
MODEL_DIR = PROJECT_ROOT / "saved_models"

MODEL_DIR.mkdir(exist_ok=True)

dataset = pd.read_csv(PROCESSED_DIR / "chat_dataset.csv")

print(dataset.shape)
dataset.head()
```

```python
all_text = list(dataset["question"]) + list(dataset["answer"])

print("Total Sentences:", len(all_text))
```

```python
tokenizer = Tokenizer(
    filters="",
    oov_token="<OOV>"
)

tokenizer.fit_on_texts(all_text)

print("Vocabulary Size:", len(tokenizer.word_index))
```

```python
with open(MODEL_DIR / "tokenizer.pkl", "wb") as f:
    pickle.dump(tokenizer, f)

print("✅ Tokenizer Saved Successfully!")
```

```python
sample = "how can i cancel my order"

sequence = tokenizer.texts_to_sequences([sample])

print(sample)
print(sequence)
```

```python
reverse_word_index = {
    index: word
    for word, index in tokenizer.word_index.items()
}

print(reverse_word_index[1])
print(reverse_word_index[2])
print(reverse_word_index[3])
```
