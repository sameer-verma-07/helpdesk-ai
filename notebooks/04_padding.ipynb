```python
import pandas as pd
import numpy as np
import pickle

from pathlib import Path

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
DATA_DIR = Path("../data/processed")

dataset = pd.read_csv(DATA_DIR / "chat_dataset.csv")

questions = dataset["question"].tolist()
answers = dataset["answer"].tolist()

print(len(questions))
```

```python
tokenizer = Tokenizer(
    filters="",
    lower=False,
    oov_token="<UNK>"
)

tokenizer.fit_on_texts(questions + answers)
```

```python
question_sequences = tokenizer.texts_to_sequences(questions)
answer_sequences = tokenizer.texts_to_sequences(answers)
```

```python
question_max = max(len(x) for x in question_sequences)
answer_max = max(len(x) for x in answer_sequences)

print("Longest Question:", question_max)
print("Longest Answer:", answer_max)
```

```python
MAX_LENGTH = 20
```

```python
encoder_input = pad_sequences(
    question_sequences,
    maxlen=MAX_LENGTH,
    padding="post"
)

decoder_input = pad_sequences(
    answer_sequences,
    maxlen=MAX_LENGTH,
    padding="post"
)
```

```python
print("Encoder Shape:", encoder_input.shape)
print("Decoder Shape:", decoder_input.shape)
```

```python
for i in range(3):
    print("Question:", encoder_input[i])
    print("Answer:", decoder_input[i])
    print("-" * 60)
```
