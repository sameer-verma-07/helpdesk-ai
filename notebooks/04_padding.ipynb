```python
import pickle
import numpy as np
import pandas as pd

from pathlib import Path

from tensorflow.keras.preprocessing.sequence import pad_sequences

print("🚀 HelpDesk AI - Padding Started")
```

```python
PROJECT_ROOT = Path.cwd().parent

PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
MODEL_DIR = PROJECT_ROOT / "saved_models"

dataset = pd.read_csv(PROCESSED_DIR / "chat_dataset.csv")

with open(MODEL_DIR / "tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

print(dataset.shape)
```

```python
question_sequences = tokenizer.texts_to_sequences(dataset["question"])

answer_sequences = tokenizer.texts_to_sequences(dataset["answer"])
```

```python
question_lengths = [len(x) for x in question_sequences]

answer_lengths = [len(x) for x in answer_sequences]

print("Max Question Length:", max(question_lengths))
print("Max Answer Length:", max(answer_lengths))
```

```python
MAX_LENGTH = 64
```

```python
encoder_input = pad_sequences(
    question_sequences,
    maxlen=MAX_LENGTH,
    padding="post",
    truncating="post"
)
```

```python
decoder_input = pad_sequences(
    answer_sequences,
    maxlen=MAX_LENGTH,
    padding="post",
    truncating="post"
)
```

```python
decoder_output = np.zeros_like(decoder_input)

decoder_output[:, :-1] = decoder_input[:, 1:]
```

```python
print("Encoder:", encoder_input.shape)
print("Decoder Input:", decoder_input.shape)
print("Decoder Output:", decoder_output.shape)
```

```python
import pickle

with open(PROCESSED_DIR / "padding.pkl", "wb") as f:
    pickle.dump(
        {
            "encoder_input": encoder_input,
            "decoder_input": decoder_input,
            "decoder_output": decoder_output,
            "max_length": MAX_LENGTH,
        },
        f,
    )

print("✅ Padding saved!")
```
